```python
import shapely
```

```python
rectangle = shapely.Polygon([(0, 0), (0, 1), (1, 1), (1, 0)])
line = shapely.LineString([(0, 0), (0, 1)])

union = rectangle.intersects(line)
```

```python
rectangle = shapely.Polygon([(0, 0), (0, 1), (1, 1), (1, 0)])
point = shapely.Point(0, 0)
line = shapely.LineString([(0, 0), (0, 0)])

rectangle.intersection(point), rectangle.intersection(line)
```

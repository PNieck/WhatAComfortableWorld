```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
import tokens
```

```python
with open("data/sequences.csv", mode="r") as file:
    line = file.readline()

    data = np.fromstring(line, dtype=np.int32, sep=',')

del file, line

data
```

# Draw boundary

```python
start, stop = np.argwhere(data == tokens.BOUNDARY_TOKEN_ID)

start = np.squeeze(start) + 1
stop = np.squeeze(stop)
```

```python
boundary = data[start:stop]
```

```python
boundary = np.reshape(boundary, (-1, 2))
```

```python
x = boundary[:, 0]
y = boundary[:, 1]

x = np.append(x, x[0])
y = np.append(y, y[0])
```

```python
plt.plot(x, y)
plt.axis('equal')
plt.grid()
plt.gca().invert_yaxis()
```
